```python
import requests  # For API call
import pandas as pd  # For handling data
```

```python
# SpaceX API endpoint for past launches
url = "https://api.spacexdata.com/v4/launches"

# Make API request
response = requests.get(url)

# Convert API response to JSON format
data = response.json()  
```

```python
# Convert JSON data to Pandas DataFrame
df = pd.json_normalize(data)

# Display first 5 rows
display(df.head())
```

```python
# Save data as a CSV file
df.to_csv("spacex_raw_data.csv", index=False)

print("✅ Data collection complete! Data saved as 'spacex_raw_data.csv'")
```

```python
# Load the raw data
df = pd.read_csv("spacex_raw_data.csv")

# Display first few rows
display(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())

# Fill missing numeric values with the median
df.fillna(df.median(numeric_only=True), inplace=True)

# Fill missing categorical values with 'Unknown'
df.fillna("Unknown", inplace=True)

print("✅ Missing values handled successfully!")
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)

print("✅ Duplicate rows removed!")
```

```python
# Convert 'date_utc' column to datetime format
df["date_utc"] = pd.to_datetime(df["date_utc"], errors="coerce")

# Convert numeric columns
df["payload_mass_kg"] = pd.to_numeric(df["payload_mass_kg"], errors="coerce")

print("✅ Data types converted!")
```

```python
# Check and rename column if needed
if "Payload Mass (kg)" in df.columns:
    df.rename(columns={"Payload Mass (kg)": "payload_mass_kg"}, inplace=True)

# Convert numeric columns safely
df["payload_mass_kg"] = pd.to_numeric(df["payload_mass_kg"], errors="coerce")

print("✅ Data types converted successfully!")
```

```python
import requests
import pandas as pd

# Function to fetch payload mass from SpaceX API
def get_payload_mass(payload_id):
    url = f"https://api.spacexdata.com/v4/payloads/{payload_id}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        return data.get("mass_kg", None)  # Return payload mass
    return None

# Extract payload mass for each row (first payload in the list)
df["payload_mass_kg"] = df["payloads"].apply(lambda x: get_payload_mass(x[0]) if isinstance(x, list) and len(x) > 0 else None)

# Convert to numeric
df["payload_mass_kg"] = pd.to_numeric(df["payload_mass_kg"], errors="coerce")

# Check results
print(df[["payloads", "payload_mass_kg"]].head(5))
```
