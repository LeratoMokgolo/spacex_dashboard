```python
import folium

# Define a map centered at SpaceX launch site
spacex_location = [28.5721, -80.6480]  # Coordinates for Kennedy Space Center, Florida
map_spacex = folium.Map(location=spacex_location, zoom_start=10)

# Display the map
map_spacex
```

```python
# List of SpaceX launch sites with coordinates
launch_sites = {
    "CCAFS LC-40": [28.5623, -80.5774],
    "VAFB SLC-4E": [34.6321, -120.6106],
    "KSC LC-39A": [28.5721, -80.6480],
    "CCAFS SLC-40": [28.5619, -80.5774]
}

# Add markers for each launch site
for site, coords in launch_sites.items():
    folium.Marker(
        location=coords,
        popup=f"Launch Site: {site}",
        icon=folium.Icon(color="blue", icon="rocket")
    ).add_to(map_spacex)

# Display the updated map
map_spacex
```

```python
# Add a circle around Kennedy Space Center
folium.Circle(
    location=spacex_location,
    radius=5000,  # Radius in meters
    color="red",
    fill=True,
    fill_color="red",
    fill_opacity=0.3
).add_to(map_spacex)

# Show the map with circles
map_spacex
```
