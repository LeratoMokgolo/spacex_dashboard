```python
import pandas as pd

# Load the dataset
df = pd.read_csv("spacex_launch_data.csv")

# Check for missing values
print(df.isnull().sum())

# Check data types
print(df.dtypes)

# Summary statistics
print(df.describe())
```

```python
import matplotlib.pyplot as plt

df["Success"].value_counts().plot(kind="bar", color=["red", "green"])
plt.title("Launch Success Distribution")
plt.xlabel("Success (0 = Failed, 1 = Successful)")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Payload Mass (kg)"])
plt.title("Payload Mass Distribution")
plt.show()
```

```python
plt.scatter(df["Payload Mass (kg)"], df["Success"], alpha=0.5)
plt.xlabel("Payload Mass (kg)")
plt.ylabel("Launch Success")
plt.title("Payload Mass vs. Success Rate")
plt.show()
```

```python
df["Launch Site"].value_counts().plot(kind="bar", color="blue")
plt.title("Launch Count per Site")
plt.xlabel("Launch Site")
plt.ylabel("Number of Launches")
plt.xticks(rotation=45)
plt.show()
```

```python
df["Success"].value_counts().plot.pie(autopct="%1.1f%%", labels=["Failure", "Success"], colors=["red", "green"])
plt.title("Proportion of Successful Launches")
plt.show()
```

```python
df["Launch Date"] = pd.to_datetime(df["Launch Date"])  # Convert to datetime
df.set_index("Launch Date").resample("Y").count()["Mission Name"].plot(kind="line", marker="o")
plt.title("Launch Frequency Over Time")
plt.xlabel("Year")
plt.ylabel("Number of Launches")
plt.show()
```
